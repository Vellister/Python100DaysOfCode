```python
import kagglehub
import shutil
from pathlib import Path

path = kagglehub.dataset_download("kartik2112/fraud-detection")
print(f"Dataset: {path}")

source_train = Path(path) / "fraudTrain.csv"
source_test = Path(path) / "fraudTest.csv"

dest_folder = Path("data/raw")
dest_folder.mkdir(parents=True, exist_ok=True)

if source_train.exists():
    shutil.copy(source_train, dest_folder / "fraudTrain.csv")
    print(f" fraudTrain.csv success in copying: {dest_folder}")

if source_test.exists():
    shutil.copy(source_test, dest_folder / "fraudTest.csv")
    print(f" fraudTest.csv success in copying: {dest_folder}")

print("\n dataset info:")
import pandas as pd
df = pd.read_csv(dest_folder / "fraudTrain.csv")
print(f"rows: {len(df):,}")
print(f"columns: {list(df.columns)}")
print(f"period: {df['trans_date_trans_time'].min()} til {df['trans_date_trans_time'].max()}")
```

```python
import pandas as pd

df_clean = df.copy()

# 1. List of columns that don't add value to our ML model
# IDs removed and personal info to avoid overfitting
cols_to_exclude = [
    'Unnamed: 0', 'cc_num', 'first', 'last', 'dob', 'trans_num',
    'merchant', 'job', 'street', 'city', 'state', 'zip'
]

# 2. Dropping the noise columns
df_clean = df_clean.drop(columns=[c for c in cols_to_exclude if c in df_clean.columns])

# 3. Time Analysis
# Converting the string timestamp into a proper datetime object
df_clean['trans_datetime'] = pd.to_datetime(df_clean['trans_date_trans_time'])
df_clean['hour'] = df_clean['trans_datetime'].dt.hour
df_clean['day_of_week'] = df_clean['trans_datetime'].dt.dayofweek
df_clean['month'] = df_clean['trans_datetime'].dt.month

# 4. Clean up temporary date columns
df_clean = df_clean.drop(columns=['trans_date_trans_time', 'trans_datetime', 'unix_time'])

# 5. categorical encoding
df_final = pd.get_dummies(df_clean, columns=['category', 'gender'], drop_first=True)

# 6. and results 
print("\n--- Processing Complete ---")
print(f"Final shape: {df_final.shape}")
df_final.info()
```

```python
# Separate features (X) and target (y)
X = df_final.drop('is_fraud', axis=1)
y = df_final['is_fraud']

print(f"Features shape: {X.shape}")
print(f"Target shape: {y.shape}")
print(f"\nTarget distribution:\n{y.value_counts()}")
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, 
    test_size=0.2,
    stratify=y,
    random_state=42
)

print(f"Train: {X_train.shape}, Test: {X_test.shape}")
print(f"Fraud in train: {y_train.sum()} ({y_train.mean()*100:.2f}%)")
print(f"Fraud in test: {y_test.sum()} ({y_test.mean()*100:.2f}%)")
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt


rf_model = RandomForestClassifier(
    n_estimators=100,
    max_depth=10, 
    class_weight='balanced',
    random_state=42,
    n_jobs=-1 
)

#  training the model
print("Training the baseline Random Forest model... (this might take a minute)")
rf_model.fit(X_train, y_train)

# 2 Making predictions
y_pred = rf_model.predict(X_test)

# 3. evaluate
print("\n--- Model Evaluation ---")
print(classification_report(y_test, y_pred))

# 4 confusion matrix
fig, ax = plt.subplots(figsize=(8, 6))
ConfusionMatrixDisplay.from_predictions(y_test, y_pred, ax=ax, cmap='Blues')
plt.title('Confusion Matrix: Baseline Random Forest')
plt.show()
```

```python
import numpy as np

# 1. feature importance
features = X.columns
importances = rf_model.feature_importances_
indices = np.argsort(importances)[-10:] # Top 10 features

plt.figure(figsize=(10, 6))
plt.title('Top 10 Feature Importances')
plt.barh(range(len(indices)), importances[indices], color='b', align='center')
plt.yticks(range(len(indices)), [features[i] for i in indices])
plt.xlabel('Relative Importance')
plt.show()

# 2. Testing a higher threshold to improve pprecision
y_probs = rf_model.predict_proba(X_test)[:, 1]
new_threshold = 0.7 
y_pred_new = (y_probs >= new_threshold).astype(int)

print(f"\n--- Evaluation with Threshold {new_threshold} ---")
print(classification_report(y_test, y_pred_new))
```

```python
from xgboost import XGBClassifier
from sklearn.metrics import classification_report, ConfusionMatrixDisplay

# 1. calculate the scale_pos_weight
# Since our data is imbalanced, XGBoost uses this parameter to 
# give more weight to the minority class (fraud).
# Formula: sum(negative cases) / sum(positive cases)
counter = y_train.value_counts()
estimate = counter[0] / counter[1]

# 2. using a lower learning_rate to make the model more robust
xgb_model = XGBClassifier(
    n_estimators=200,
    max_depth=6,
    learning_rate=0.1,
    scale_pos_weight=estimate,
    random_state=42,
    use_label_encoder=False,
    eval_metric='logloss',
    n_jobs=-1
)

# 3. train the model
print(f"Training XGBoost with scale_pos_weight={estimate:.2f}...")
xgb_model.fit(X_train, y_train)

# 4. oredict with a threshold of 0.7 (to compare with our previous result)
xgb_probs = xgb_model.predict_proba(X_test)[:, 1]
xgb_pred_07 = (xgb_probs >= 0.7).astype(int)

# 5. evaluate
print("\n--- XGBoost Evaluation (Threshold 0.7) ---")
print(classification_report(y_test, xgb_pred_07))

# 6. plot Confusion Matrix
fig, ax = plt.subplots(figsize=(8, 6))
ConfusionMatrixDisplay.from_predictions(y_test, xgb_pred_07, ax=ax, cmap='Greens')
plt.title('Confusion Matrix: XGBoost (Threshold 0.7)')
plt.show()
```

```python
import joblib
from pathlib import Path

model_path = Path("../data/models")
model_path.mkdir(parents=True, exist_ok=True)
joblib.dump(xgb_model, model_path / "xgboost_fraud_model.pkl")
joblib.dump(list(X.columns), model_path / "model_features.pkl")

print(f" Model and features saved successfully in: {model_path}")
```

```python
from sklearn.metrics import roc_auc_score, average_precision_score

roc_auc = roc_auc_score(y_test, xgb_probs)

pr_auc = average_precision_score(y_test, xgb_probs)

print(f"--- Advanced Metrics ---")
print(f"ROC-AUC Score: {roc_auc:.4f}")
print(f"PR-AUC (Average Precision): {pr_auc:.4f}")
```

```python
false_positives = X_test[(y_test == 0) & (xgb_pred_07 == 1)].copy()
false_negatives = X_test[(y_test == 1) & (xgb_pred_07 == 0)].copy()

print(f"--- Error Statistics ---")
print(f"False Positives (FP): {len(false_positives)} instances")
print(f"False Negatives (FN): {len(false_negatives)} instances")

print("\n--- Insight: Why did it fail???? ---")
print(f"Average 'amt' of False Positives: ${false_positives['amt'].mean():.2f}")
print(f"Average 'amt' of False Negatives: ${false_negatives['amt'].mean():.2f}")

error_report_path = "../reports/error_analysis.csv"
pd.concat([false_positives.assign(error_type='FP'), 
           false_negatives.assign(error_type='FN')]).to_csv(error_report_path)

print(f"\n Error report saved to: {error_report_path}")
```
